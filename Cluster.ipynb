```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
from sklearn.cluster import KMeans
```

```python
# Importing the dataset
df = pd.read_excel (r'C:\Users\mjanardanan\Desktop\Clustering\Custom SQL Query_thurs.xlsx') 
df.head(10)
```

```python
#f1 = df['Cnt'].values
f2 = df['EntBin'].values
f3 = df['losn'].values
#f4 = df['Cnt'].values
X = np.array(list(zip(f2, f3)))
X
```

```python


# Number of clusters
kmeans = KMeans(n_clusters=14)
# Fitting the input data
kmeans = kmeans.fit(X)
# Getting the cluster labels
labels = kmeans.predict(X)
# Centroid values
centroids = kmeans.cluster_centers_
#print(labels)
print(centroids)
np.savetxt('fullthurs.txt', labels, fmt='%.f')
```
